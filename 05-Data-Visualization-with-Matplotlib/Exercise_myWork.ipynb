```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('title')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])

ax1.plot(x,y,color='r')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

ax2.plot(x,y,color='r')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])

ax1.plot(x,z,color='b')
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

ax2.plot(x,y,color='b')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y,color='blue',lw=2,ls='--')
axes[1].plot(x,z,color='red',lw=2,ls='-')

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(8,2))

axes[0].plot(x,y,color='blue',lw=2,ls='--')
axes[1].plot(x,z,color='red',lw=2,ls='-')

plt.tight_layout()
```
