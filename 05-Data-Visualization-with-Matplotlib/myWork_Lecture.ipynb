```python
# OO Object-Oriented
# create figure object and call methods off of it
fig = plt.figure() #Imaginary blank canvas
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
type(fig)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
#Plot appearance
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00') # RGB Hex Code

ax1 = fig.add_axes([0.2,0.2,0.5,0.5])
ax1.plot(y,x,color='green')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Smaller Graph')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red', linewidth=20)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange', lw=20, alpha=0.50)
```
