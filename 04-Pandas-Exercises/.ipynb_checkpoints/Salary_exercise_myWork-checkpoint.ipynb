```python
#How many people have the word Chief in their job title? (This is pretty tricky)
def check_substring(word):
    if "chief" in word.lower():
        return True
    else:
        return False
```

```python
#How many people have the word Chief in their job title? (This is pretty tricky)
sal['JobTitle'][sal['JobTitle'].apply(check_substring) == True]
```

```python
#How many people have the word Chief in their job title? (This is pretty tricky)
sal['JobTitle'][sal['JobTitle'].apply(check_substring) == True].count()
```
