```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.jointplot(x='fare',y='age',data=titanic,xlim=(-100,600))
```

```python
sns.distplot(titanic['fare'],kde=False,bins=30,color='r')
```

```python
titanic
```

```python
sns.boxplot(x='class',y='age',data=titanic,palette='rainbow')
```

```python
sns.swarmplot(x='class',y='age',data=titanic,palette='Set2')
```

```python
sns.countplot(titanic['sex'])
```

```python
tn = titanic.corr()
sns.heatmap(tn,cmap='coolwarm').set_title('titatic.corr()')
```

```python
plot = sns.FacetGrid(data=titanic,col='sex')
plot.map(sns.distplot,'age',kde=False,bins=10)
```
