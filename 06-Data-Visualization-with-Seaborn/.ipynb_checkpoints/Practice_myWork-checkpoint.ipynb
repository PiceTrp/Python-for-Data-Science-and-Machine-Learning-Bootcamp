```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diamonds = sns.load_dataset('diamonds')
planets = sns.load_dataset('planets')
```

```python
diamonds
```

```python
planets
```

```python
diamonds['cut'].unique()
```

```python
diamonds['clarity'].unique()
```

```python
plt.figure(figsize=(12,5))
sns.set_style('ticks')
sns.set_context('notebook',font_scale=1.5)
sns.distplot(diamonds['price'],bins=100,color='g')
```

```python
sns.set_context("notebook")
sns.jointplot(x='depth',y='price',data=diamonds,kind='scatter',xlim=(55,57),height=10)
```

```python
plt.figure(figsize=(12,5))
sns.set_context('talk')
sns.violinplot(x='cut',y='price',data=diamonds)
```

```python
plt.figure(figsize=(10,5))
sns.set_context('notebook',font_scale=1.25)
sns.jointplot(x='x',y='price',data=diamonds,xlim=(7,8),ylim=(12500,13000),color='purple')
```

```python
plot = sns.FacetGrid(data=diamonds,col='cut',row='clarity',xlim=(0,10000),height=8)
plot.map(sns.distplot,'price',color='r')
```

```python
sns.jointplot(x='orbital_period',y='mass',data=planets,kind='hex',xlim=(0,2500),ylim=(0,5),color='orange')
```

```python
sns.pairplot(planets)
```

```python
sns.jointplot(x='year',y='mass',data=planets,kind='kde',color='r',xlim=(2000,2015),ylim=(-2,5))
```

```python
dots = sns.load_dataset('dots')
dots
```
