# Grid

```python
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
iris
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
sns.PairGrid(iris)
```

```python
import matplotlib.pyplot as plt
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g= sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
#FacetGrid
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.kdeplot,'total_bill','tip')
```

# Regression Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',row='time',hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.6,height=8)
```

# Style and Color

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
sns.despine(left=True,bottom=True)
```

```python
#Seaborn are all just basic callbacks to matplotlib
#So sns can override to plt

#Size and aspect
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python
#Scale and Context
sns.set_context('poster')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('poster',font_scale=0.75)
sns.countplot(x='sex',data=tips)
```

```python
#palettes and colors
sns.set_context('notebook',font_scale=1)
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='seismic')
```
